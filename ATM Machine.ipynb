```python
class Atm:
    data={'U1':['1234',21000,'user1'],'U2':['7890',42000,'user2']}
    def __init__(self):
        user=input('Enter username:\n')
        if user in self.data:
            self.name=user
            self.pin=self.data[user][0]
            self.bal=self.data[user][1]
            self.pas=self.data[user][2]
            pas=input('Enter your password')
            if pas==self.pas:
                choise=int(input('1.New user 2.Existing User\n'))
                if choise == 1:
                    self.create()
                elif choise == 2:
                    ch=int(input('1.Change Pin 2. Deposit 3. Withdraw 4. Show Balance\n'))
                    if ch == 1:
                        self.change()
                    elif ch == 2:
                        self.deposit()
                    elif ch == 3:
                        self.withdraw()
                    elif ch == 4:
                        self.show()
                    elif ch == 5:
                        exit()
                    else:
                        print('Enter correct choise..')
                else:
                    print('Enter valid choise..')
            else:
                print('Enter correct password')
        else:
            print('Enter valid username')
    def create(self):
        create_pin=input('Enter the new pin:\n')
        if len(create_pin) == 4:
            if create_pin.isdigit()==True:
                self.pin=create_pin
                print('Your new pin is generated successfully')
                self.data.update({self.name:[self.pin,self.bal]})
            else:
                print('Enter pin in digits...')
        else:
            print('Enter sufficient length..')
    def change(self):
        oldpin=input('Enter your old pin\n')
        if oldpin==self.data[self.name][0]:
            newpin=input('Enter New pin\n')
            if len(newpin) == 4:
                if newpin.isdigit()==True:
                    self.pin=newpin
                    print('Your pin is changed successfully')
                    self.data.update({self.name:[self.pin,self.bal]})
                else:
                    print('Enter pin in digits...')
            else:
                print('Enter sufficient length..')
        else:
            print('Incorrect pin')
    def pri(self):
        choose=int(input('Do you want to show balance? 1-Yes'))
        if choose==1:
            print('Your currunt balance is',self.bal)
        else:
            print('Thank you for visit..')
    def withdraw(self):
        askpin=input('Enter your pin\n')
        if askpin==self.data[self.name][0]:
            wdr=int(input('Enter amount you want to withdraw\n'))
            if wdr<=self.bal-2000:
                self.bal=self.bal-wdr
                print('The amound',wdr,'is withdraw from account')
                self.data.update({self.name:[self.pin,self.bal]})
                self.pri()
            else:
                print('Enter sufficient balance..')
        else:
            print('Incorrect pin')
    def deposit(self):
        askpin=input('Enter your pin\n')
        if askpin==self.data[self.name][0]:
            dep=int(input('Enter amount you want to Deposit\n'))
            self.bal=self.bal+dep
            self.data.update({self.name:[self.pin,self.bal]})
            self.pri()
        else:
            print('Incorrect pin')
    def show(self):
        askpin=input('Enter your old pin\n')
        if askpin==self.data[self.name][0]:
            return self.bal
        else:
            print('Incorrect pin')
        
Bank=Atm()
```
